```
class myClass_1:
  var_1 = 'Class Member'
  def func_1(self):
    var_2 = 'Instance Member'
    print('Method 내에 선언', var_2)
    print(' Class 내에 선언', self.var_1)

  def func_2(self, x, y):
    z = x + y
    print(x, '더하기', y, '는', z)
```

```
obj_1 = myClass_1()
```

```
obj_1.var_1
```

```
obj_1.var_2
```

```
obj_1.func_1()
```

```
obj_1.func_2(3, 5)
```

```
class myClass_2:
  def __init__(self):
    print('myClass_2 인스턴스가 메모리에 생성되었습니다')
    
  var_1 = 'Class Member'

  def func_1(self):
    print('Class Method')

  def __del__(self):
    print('myClass_2 인스턴스가 메모리에서 소멸되었습니다')
```

```
obj_2 = myClass_2()
```

```
obj_2.var_1
```

```
obj_2.func_1()
```

```
del obj_2
```

```
class Sum:
  def add(self, x, y):
    z = x+y
    return z

class Cop:
  def mul(self, x, y):
    z = x*y
    return z

class Computer(Sum, Cop):
  def sub(self, x, y):
    z=x-y
    return z
```

```
obj_3 = Computer()
```

```
obj_3.sub(3, 5)
```

```
obj_3.add(3,5)
```

```
obj_3.mul(3, 5)
```
