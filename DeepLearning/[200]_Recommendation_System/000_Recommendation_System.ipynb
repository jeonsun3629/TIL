# 추천 시스템(Recomendation System)

* 사용자 본인도 잘 몰랐던 취향이나 관심사를 추천 시스템이 발견하여 제시



## 관련 데이터

1. 사용자가 어떤 상품을 구매했는가?
2. 사용자가 어떤 상품을 둘러보았는가?
3. 사용자가 어떤 상품을 장바구니에 넣었는가?
4. 사용자가 상품에 어떤 평점을 부여했는가?
5. 사용자가 가입 시 선택한 관심분야는 무엇인가?

# I. 협업 필터링(Collaborative Filtering)

* 사용자 행동양식(User Behavior) 데이터
 - 사용자(User)가 아이템(Item)에 매긴 <font color = 'blue'>평점정보</font>나 상품구매이력
* 최근접이웃(Nearest Neighbor) 방식과 잠재요인(Latent Factor) 방식
 - 두 방식 모두 '사용자-아이템 행렬' 사용

> ## 1) 사용자-아이템 행렬(User-Item Matrix)

* 축적된 '사용자 행동양식(User Behavior)' 데이터 기반
* 사용자가 아직 평가하지 않은 아이템을 '<font color = 'red'>예측 평가</font>(Predicted Rating)'

||Item1|Item2|Item3|Item4|
|-|:--:|:---:|:---:|:---:|
|User1|4||3|<font color = 'red'>PR|
|User2|5|3||4|
|User3||1|2|3|

# II. 최근접 이웃 협업 필터링(Memory 협업 필터링)

* 사용자 기반과 아이템 기반으로 분류

> ## 1) 사용자 기반 필터링(User Based Filtering)

## **당신과 비슷한 고객들이 다음과 같은 상품도 구매했습니다**
* 사용자와 유사한 다른 사용자를 Top-N으로 선정
* Top-N 사용자가 좋아하는 아이템을 추천하는 방식
 1. 특정 사용자와 다른 사용자 간의 '<font color = 'red'>유사도</font>(Similarity)'를 측정
 2. 가장 유사도가 높은 Top-N 사용자를 추출
 3. 그들이 선호하는 아이템을 사용자에게 추천

# User-Item Matrix

||Item1|Item2|Item3|Item4|Item5|
|-|:--:|:---:|:---:|:---:|:---:|
|<font color = 'red'>User1|5|4|4|<font color = 'blue'>PR|<font color = 'blue'>PR|
|<font color = 'red'>User2|5|3|4|5|3|
|User3|4|3|3|2|5|

* User1과 User2의 'Item'에 대한 평점 유사도가 높음
* User1이 아직 구매하지 않은 Item4와 Item5를 추천(<font color = 'blue'>Predicted Rating</font>)

> ## 2) 아이템 기반 필터링(Item Based Filtering)

## **이 상품을 구매한 다른 고객들은 다음과 같은 상품도 구매했습니다**
* 사용자의 아이템에 대한 평가 결과가 유사한 아이템을 추천
* 행과 열이 사용자 기반 필터링과 반대
 1. 아이템에 대한 평점 간 '<font color = 'red'>유사도</font>(Similarity)' 측정
 2. 평점 유사도가 높은 아이템를 추출
 3. 그들이 선호하는 아이템을 사용자에게 추천

# Item-User Matrix

||User1|User2|User3|User4|User5|
|-|:--:|:---:|:---:|:---:|:---:|
|<font color = 'red'>Item1|5|4|5|5|5|
|<font color = 'red'>Item2|5|4|4|<font color = 'blue'>PR|5|
|Item3|4|2|2||4|

* Item1과 Item2의 평점 유사도가 높음
* User4가 아직 구매하지 않은 Item2를 추천(<font color = 'blue'>Predicted Rating</font>)

# III. 잠재요인 협업 필터링

### User-Item Matrix를 사용하여 '<font color = 'red'>잠재요인</font>(Latent Factor)'을 추출
- '잠재요인'이 무엇인지 명확하게 정의할 수 없음
- User-Item Matrix를 '잠재요인' 기반의 저차원 밀집행렬인 '사용자-잠재요인'과 '잠재요인-아이템'으로 행렬 분해(Matrix Factorization)
   - SVD(Singular Vector Decomposition)
   - NMF(Non-Negative Matrix Factorization)
- 이 두 행렬의 내적을 통해서 새로운 '**사용자-아이템 평점**' 행렬 생성
- 사용자가 평점을 부여하지 않은 아이템에 대한 '<font color = 'blue'>예측평점</font>' 생성 가능

> ## 1) User-Item Matrix

* 희소행렬(Sparse Matrix)

||Item1|Item2|Item3|Item4|Item5|
|-|:--:|:---:|:---:|:---:|:---:|
|User1|5| |4| |5|
|User2| | | |5|3|
|User3|4| | |2| |
|User4| |2| | |4|

* 확률적 경사하강(Stochastic Gradient Descent) 기반 행열 분해
 - P와 Q 행렬로 계산된 예측 R 행렬의 값이 실제 R 행렬의 값과 가장 최소의 오류를 가질 수 있도록 반복적으로 'Cost Function' 최적화를 통해 P와 Q를 유추해내는 것

1. 임의의 값을 갖는 P와 Q 행렬 초기화
2. P와 Q 행렬로 예측 R 행렬 계산 후 실제 R 행렬과 오차값 계산
3. 오차를 최소화 하도록 P와 Q 행렬의 값을 업데이트
4. 2단계와 3단계를 반복하며 P와 Q 행렬값을 학습
5. 행렬 분해가 완료된 P와 Q 행렬로 예측평점 행렬 생성

> ## 2) User-'<font color = 'red'>Latent Factor</font>' Matrix

||Factor1|Factor2|
|-|:--:|:---:|
|User1|5|4|
|User2|5|3|
|User3|4|3|
|User4|3|2|

* 가격, 브랜드, 장르 등으로 추정
